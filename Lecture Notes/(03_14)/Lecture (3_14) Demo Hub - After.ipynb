```python
dice = np.arange(6)+1
dice
```

```python
dice_outcomes = make_array()

for first_die in dice:
    trial = make_array(first_die)
    
    for second_die in dice:
        trial = np.append(trial,second_die)
        
        for third_die in dice:
            trial =  np.append(trial,third_die)
            
            for fourth_die in dice:
                trial = np.append(trial,fourth_die)
                
                dice_outcomes = np.append(dice_outcomes,trial)

dice_outcomes.item(0)
#this got way out of hand, will fix this for Thursday
```

```python
outcomes = 0
dice_rolls = 0
#the for-loops below simulate the rolling of my 4 dice.
for first_die in dice: 
    for second_die in dice:
        for third_die in dice:
            for fourth_die in dice:
                
                #the array created below stores my outcomes for the four dice. 
                sample_outcome = make_array(first_die,second_die,third_die,fourth_die)
                
                #since I rolled my 4 dice increase the count of dice rolls by one
                dice_rolls += 1 # is the same as: (dice_rolls = dice_rolls + 1) i.e. increases the count by 1
                
                #check to see if there is at least one '6'
                if np.count_nonzero(sample_outcome == 6) >= 1: #True if there is at least one '6' in the sample outcome array
                    outcomes += 1
                
                print(sample_outcome)
print(outcomes)
print(dice_rolls)
```
