# Investigating Swain v. Alabama

```python
black_nonblack_proportion_array = make_array(.26,.74)
```

```python
sample_proportions(100,black_nonblack_proportion_array)*100
```

```python
simulated_juries = make_array()
num_simulations = 10000

for i in np.arange(num_simulations):
    jury_selection = sample_proportions(100,black_nonblack_proportion_array)*100
    num_of_black_jurors = jury_selection.item(0)
    simulated_juries = np.append(simulated_juries,num_of_black_jurors)

simulated_juries
```

```python
t = Table().with_column('Simulated Juries', simulated_juries)
t.hist()
plt.scatter(8, 0, color='red', s=100, zorder=2);
```
